### Best Way To Separate the Dataframe I've Found

```python
pitch_type_woba = first_ab.groupby('pitch_type', as_index=False).agg({"woba_value": "sum", "woba_denom" : "sum",\
                                                                         "estimated_woba_using_speedangle" : "mean"})

pitch_type_woba['Actual wOBA'] = pitch_type_woba['woba_value'] / pitch_type_woba['woba_denom']
pitch_type_woba['wOBA Differential'] = pitch_type_woba['estimated_woba_using_speedangle']\
                            - pitch_type_woba['Actual wOBA']
    
pitch_type_woba = pitch_type_woba.drop(columns=['woba_value'])
pitch_type_woba = pitch_type_woba.rename(columns={"pitch_type":"Pitch", \
                                                    "estimated_woba_using_speedangle" : "Estimated wOBA",\
                                                 "woba_denom" : "Plate Appearances Ending with Pitch"})

cols = list(pitch_type_woba.columns.values)
final_cols = ['Pitch',
              'Plate Appearances Ending with Pitch',
              'Estimated wOBA',
              'Actual wOBA',
              'wOBA Differential']

final_pitch_type_woba = pitch_type_woba[final_cols]
final_pitch_type_woba = final_pitch_type_woba.round(3)
final_pitch_type_woba
```

```python
pitch_type_woba2 = rest_of_abs.groupby('pitch_type', as_index=False).agg({"woba_value": "sum", "woba_denom" : "sum",\
                                                                         "estimated_woba_using_speedangle" : "mean"})

pitch_type_woba2['Actual wOBA'] = pitch_type_woba2['woba_value'] / pitch_type_woba2['woba_denom']
pitch_type_woba2['wOBA Differential'] = pitch_type_woba2['estimated_woba_using_speedangle']\
                            - pitch_type_woba2['Actual wOBA']
    
pitch_type_woba2 = pitch_type_woba2.drop(columns=['woba_value'])
pitch_type_woba2 = pitch_type_woba2.rename(columns={"pitch_type":"Pitch", \
                                                    "estimated_woba_using_speedangle" : "Estimated wOBA",\
                                                 "woba_denom" : "Plate Appearances Ending with Pitch"})

cols = list(pitch_type_woba2.columns.values)
final_cols = ['Pitch',
              'Plate Appearances Ending with Pitch',
              'Estimated wOBA',
              'Actual wOBA',
              'wOBA Differential']

final_pitch_type_woba2 = pitch_type_woba2[final_cols]
final_pitch_type_woba2 = final_pitch_type_woba2.round(3)
final_pitch_type_woba2
```

```python
def pos_green_neg_red(val):
    color = 'red' if val < 0 else 'green'
    return 'background-color: %s' % color
```

```python
def woba_values(val):
    if val >= .400:
        color = 'green'
    elif .400 > val >= .370:
        color = '#66ff33'
    elif .370 > val >= .340:
        color = '#b3ff99'
    elif .340 > val >= .320:
        color = '#gray'
    elif .320 > val >= .310:
        color = '#FFFF00'
    elif .310 > val >= .300:
        color = '#ff6600'
    else:
        color = 'red'
    return 'background-color: %s' % color
```

```python
fab = final_pitch_type_woba.style \
                        .applymap(pos_green_neg_red, subset=pd.IndexSlice[:, ['wOBA Differential']])\
                        .applymap(woba_values, subset=pd.IndexSlice[:, ['Estimated wOBA', 'Actual wOBA']])
fab
```

```python
rob = final_pitch_type_woba2.style \
                        .applymap(pos_green_neg_red, subset=pd.IndexSlice[:, ['wOBA Differential']])\
                        .applymap(woba_values, subset=pd.IndexSlice[:, ['Estimated wOBA', 'Actual wOBA']])
rob
```
